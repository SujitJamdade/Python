```python
class Container:
    def __init__(self, id,length,breadth, height, pricePerSqrFt):
        self.id = id
        self.le = length
        self.br = breadth
        self.he = height
        self.pricePerSqrFt = pricePerSqrFt
        
    def findVolume(self):
        return self.le * self.br * self.he
           
class PackagingCompany:
    
    def __init__(self,containerList):
        self.cl = containerList
        
    def findContainerCost(self,id1):
        
        for i in self.cl:
            
            if i.id == id1:
                
                volume_container = i.findVolume()
                
                return volume_container * i.pricePerSqrFt
        
        return "No container found"
        
    def findLargestContainer(self):
        result_volume = []
        id_out = []
        for i in self.cl:
            result_volume.append(i.findVolume())
            id_out.append(i.id)
            
        out1 = max(result_volume)
        output1=[]
        for  i in result_volume:
            if out1==i:
                index1 =  id_out[result_volume.index(out1)] 
                output1.append(index1)
        
        output1.append(out1)
        
        return output1     
    
if __name__ == "__main__":
    N = int(input())
    containerList = []
    for i in range(N):
        id = int(input())
        length = int(input())
        bredth = int(input())
        height = int(input())
        price = int(input())
        containerList.append(Container(id,length,bredth,height,price))
        
    obj1 = PackagingCompany(containerList)
    id1 = int(input())
    
    print(obj1.findContainerCost(id1))
    
    result1 = obj1.findLargestContainer()
    
    for i in result1:
        print(i, end = " ")   
```
